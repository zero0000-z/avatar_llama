```python
folder_path = '/code/Video-LLaMA-main/data/zoo_300k/zoo_300k/zoo_300k/AnacondaALL'
def get_filepath(folder_path):
    path_list = []
    for file in os.listdir(folder_path):
        path_list.append(os.path.join(folder_path, file))
    return path_list
```

```python
def get_mp4_files(directory):
    mp4_files = []
    for root, dirs, files in os.walk(directory):
        first_mp4_found = False
        mp4s_in_dir = [file for file in files if file.endswith('.mp4')  and not 'input' in file and not 'sample' in file]
        if mp4s_in_dir:
            for mp4 in mp4s_in_dir:
            # Only add one mp4 file per directory
                mp4_files.append(os.path.join(root, mp4))
            # break  # Only take the first .mp4 file in each subfolder
        # if first_mp4_found:
        #     break  # Once a .mp4 file is found in the current folder, skip to next folder
    return mp4_files

# 指定文件夹路径
folder_path = '/code/Video-LLaMA-main/data/zoo_300k/zoo_300k/zoo_300k/AnacondaALL'
filelist = get_mp4_files(folder_path)
filelist[:3]
```

```python
import os

def find_deepest_folders_with_mp4(root_dir):
    def is_mp4_present_in_folder(folder_path):
        return any(f.endswith('.mp4') for f in os.listdir(folder_path))
    
    deepest_folders = []
    max_depth = 0
    
    for root, dirs, files in os.walk(root_dir):
        depth = root[len(root_dir):].count(os.sep)
        if depth > max_depth:
            max_depth = depth
            deepest_folders = [root]
        elif depth == max_depth:
            deepest_folders.append(root)
    
    for folder in deepest_folders:
        if is_mp4_present_in_folder(folder):
            print(f'MP4 file found in: {folder}')
        else:
            print(f'No MP4 file in: {folder}')

# Example usage

folder_path = '/code/Video-LLaMA-main/data/zoo_300k/zoo_300k/zoo_300k/AnacondaALL'
filelist = find_deepest_folders_with_mp4(folder_path)
filelist
```

```python
def read_json(filepath):
    with open(filepath, "r",encoding="utf-8") as f:
        data= json.load(f)
    return data


def write_json(data, file_path):
    with open(file_path, 'w') as f:
        json.dump(data, f, indent=4)

# write_json(filelist, 'data/horse_all.json')
```

```python
data = read_json('zoo_300k/zoo_300k_llama13B.json')
```

```python
data_all = {}
for line in data[:3]:
    filepath = line['filepath']
    animal, action = re.split('/', filepath)[7:9]


    if data_all is None or animal not in data_all.keys():
        content = []
        content.append(line)
        animal_dict = [{"action":action,
                       "content":content
                       }]
        data_all[animal] = animal_dict

    else:
        content  = data_all[animal]
        animal_dict = {"action":action,
                "content":content
                }
        content.append(animal_dict)

        data_all[animal]=content

```
